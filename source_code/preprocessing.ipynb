```python
import pandas as pd
import numpy as np
import re
```

```python
data = pd.read_csv("../data/reviews_with_category.csv", index_col=0)
```

```python
data = data[['platform', 'store', 'date', 'review', 'rating']]
data
```

## 어미가 '요'로 끝나면 다른 행으로 리뷰 끊기

```python
yo_li = ['요 ', '요ㅋ', '요ㅎ', '요!', '요?', '요.', '요>', '요ㅡ', '요ㅜ', '요ㅠ']

test = data['review'][1].replace('요', '요요')
test.split('요.')

# for idx, row in data.iterrows():
```

```python
for idx, row in data.iterrows():
    for yo in yo_li:
        if row['review'].find(yo) != -1:
            data['review'][idx] = row['review'].replace(yo, '요오,.')

data
```

```python
for idx, row in data.iterrows():
    data['review'][idx] = row['review'].split('오,')

data
```

```python
data = data.explode('review', ignore_index=True)
data
```

## 정규 표현식으로 한글이 아니면 제거

```python
for idx, row in data.iterrows():
    data['review'][idx] = re.sub('[^가-힣ㄱ-ㅎㅏ-ㅣ\\s]','',row.review)
data
```

## 리뷰 앞에 공백 제거

```python
for idx, row in data.iterrows():
    if row.review[:1] == ' ':
        data['review'][idx] = row.review[1:]

data
```

## 어미가 '니다'로 끝나면 다른 행으로 리뷰 끊기

```python
for idx, row in data.iterrows():
    if row['review'].find('니다') != -1:
        data['review'][idx] = row['review'].replace('니다', '니다오.')

data
```

```python
for idx, row in data.iterrows():
    data['review'][idx] = row['review'].split('오.')

data
```

```python
data = data.explode('review', ignore_index=True)
data
```
