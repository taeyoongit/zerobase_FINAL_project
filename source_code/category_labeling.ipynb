```python
# 공백행 제거하기

del_index = []

for idx, row in data.iterrows():
    if data['review'][idx] == '':
        del_index.append(idx)

data = data.drop(index=del_index)
data
```

```python
# 공백행 제거하기2

del_index = []

for idx, row in data.iterrows():
    if data['review'][idx] == ' ' or data['review'][idx] == '  ' or data['review'][idx] == ' ' or data['review'][idx] == '   ':
        del_index.append(idx)

data = data.drop(index=del_index)
data
```

```python
# 현재 기준으로 카테고리 다시 주기

data['category'] = None

for idx, row in data.iterrows():
    temp = []
    for key, value in category_dic.items():
        for val in value:
            if row['review'].find(val) != -1:
                temp.append(key)

    if temp != []:
        new_temp = []
        for v in temp:
            if v not in new_temp:
                new_temp.append(v)
        data['category'][idx] = new_temp

for i, r in data.iterrows():
    if r.category != None:
        data['category'][i] = ', '.join(r.category)
    else:
        data['category'][i] = '만족도&기타'

data
```
